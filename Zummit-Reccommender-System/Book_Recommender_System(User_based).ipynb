```python
#Importingthe neccessary libaries
import numpy as np
import pandas as pd
```

```python
#importing the dataset and converting it into a dataframe
book_df= pd.read_csv('Books.csv', low_memory=False)
ratings_df = pd.read_csv('Ratings.csv')
users_df = pd.read_csv('Users.csv')
```

```python
#Checking the first five datapoints
book_df.head()
```

```python
#Removing the redundant and unneccessary columns
book_df.drop(['Book-Author', 'Year-Of-Publication', 'Publisher','Image-URL-S','Image-URL-M'],
            inplace= True, axis=1)
print(book_df.head())
```

```python
#Checking the information details of the dataset
book_df.info()
```

```python
#Checking the information details of the ratings dataset
print(ratings_df)
print('\n')
ratings_df.info()
```

```python
#Checking the first 5 datapoints of the user dataset
users_df.head()
```

```python
#Checking the information details of the user dataset
users_df.info()
```

```python
#Selecting the users that rated books more than 200 times
ratings_200 = ratings_df['User-ID'].value_counts() > 200
```

```python
rating_index = ratings_200[ratings_200].index
rating_index
```

```python
#Selecting the datapoints based on the users whom rated over 200 books and making it the main dataset that I'll be working with
ratings = ratings_df[ratings_df['User-ID'].isin(rating_index)]
```

```python
ratings
```

```python
#Merging the ratings and book_df dataset on ISBN column
df = ratings.merge(book_df, on='ISBN')
df
```

```python
#Creating a dataset where it is grouped by the book title and the number of times it was rated
number_rating = df.groupby('Book-Title')['Book-Rating'].count().reset_index()
number_rating.head()
```

```python
number_rating.rename(columns={'Book-Rating':'num_of_rating'},inplace=True)
number_rating.head()
```

```python
#Merging the number_rating with our main dataset, df, 
final_rating = df.merge(number_rating, on='Book-Title')
final_rating.head()
```

```python
# Lets take those books which got at least 50 rating of user

final_rating = final_rating[final_rating['num_of_rating'] >= 50]
final_rating.head()
```

```python
# lets drop the duplicates
final_rating.drop_duplicates(['User-ID','Book-Title'],inplace=True)
```

```python

# Lets create a pivot table, where the columns is the User ID and Book Title represents the row 
book_pivot = final_rating.pivot_table(columns='User-ID', index='Book-Title', values= 'Book-Rating')
book_pivot
```

```python
#Filling the NaN with zeros
book_pivot.fillna(0, inplace=True)
book_pivot
```

```python
#Converting the book_pivot into a sparse martix
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
# Now import our clustering algoritm which is Nearest Neighbors this is an unsupervised learning model
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python
#Using the model to get the distance between book liked by the user and the 6 closest books to it
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6 )
```

```python
print(distance, suggestion, sep='\n')
```

```python
#Defining a function for the recommendation
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
```

```python
recommend_book("Harry Potter and the Chamber of Secrets (Book 2)")
```
